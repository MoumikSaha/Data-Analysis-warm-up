```python
def AvgMarks(x):
  return x/2
```

```python
df1["Half marks"] = df1['Marks'].apply(AvgMarks) #using user define function
df1
```

```python
df1["Half marks"] = df1['Marks'].apply(lambda x:x/2) #Using Lambda Functions
df1
```
